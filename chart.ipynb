```python
import matplotlib.pyplot as plt

# data
adv = [56.82, 54.35, 52.77, 52.17, 52.09]
std = [39.27, 0.77, 0.02, 0.0, 0.0]
x_axis = [1, 5, 10, 20, 50]

# plot lines
plt.plot(x_axis, adv, label='Adv')
plt.plot(x_axis, std, label='Std')

# plot data points
plt.scatter(x_axis, adv)
plt.scatter(x_axis, std)

# set labels and title
plt.xlabel('PGD Attack Iterations')
plt.ylabel('Accuracy (%)')
plt.title('Comparison of Adv and Std with iteration change')

# set legend
plt.legend()

# display the plot
plt.show()

```
